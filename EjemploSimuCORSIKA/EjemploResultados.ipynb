# Resultados para ejemplo de simulación en CORSIKA

Este Notebook de ejemplo muestra los resultados de una simulación de ratos cósmicos con CORSIKA hecha para la ciudad de Montería, a 18 msnm. El tiempo de flujo fue de 360 segundos.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Primero usaremos el archivo salidasimu.dst, el cual es un histograma de los resultados donde las partículas detectadas se ordenan por la distancia radial desde el centro de la lluvia a la cual fueron detectadas.

```python
namesDST = ["distance_in_bin(m)", "N_phot", "N_e+", "N_e-", "N_mu+", "N_mu-", "N_pi0", "N_pi+", "N_pi-", "N_n", "N_p", "N_pbar", "N_others", "Total_per_bin"]
dst = pd.read_csv("salidasimu.dst", skiprows=6, skipfooter=7, delimiter="\s+", names=namesDST)
```

```python
dst
```

A continuación, está una figura donde se muestra el número de partículas detectadas en función de la distancia al centro de la lluvia. Se muestran las curvas para fotones, electrones más positrones, muones + más muones -, piones y protones más neutrones, así como también para todas las partículas juntas.

```python
plt.figure(figsize=(10,6))
plt.plot(dst["distance_in_bin(m)"], dst["N_phot"], label="Fotones")
plt.plot(dst["distance_in_bin(m)"], dst["N_e+"] + dst["N_e-"], label="Electrones y positrones")
#plt.plot(dst["distance_in_bin(m)"], dst["N_e-"], label="N electrons")
plt.plot(dst["distance_in_bin(m)"], dst["N_mu+"] + dst["N_mu-"], label="Muones + y -")
#plt.plot(dst["distance_in_bin(m)"], dst["N_mu-"], label="N muons-")
plt.plot(dst["distance_in_bin(m)"], dst["N_pi+"] + dst["N_pi-"] + dst["N_pi0"], label="Piones +, - y 0")
plt.plot(dst["distance_in_bin(m)"], dst["N_p"] + dst["N_n"], label="Protones y neutrones")
plt.plot(dst["distance_in_bin(m)"], dst["Total_per_bin"], label="Total")

plt.xlabel("Distancia (m)")
plt.ylabel("Número de partículas")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.grid()
plt.show()
```

Ahora observaremos el archivo salidasimu.hst, en el que las secundarias han sido ordenadas por su energía.

```python
namesHST = ["p_in_bin(GeV)", "N_phot", "N_e+", "N_e-", "N_mu+", "N_mu-", "N_pi0", "N_pi+", "N_pi-", "N_n", "N_p", "N_pbar", "N_others", "Total_per_bin"]
hst = pd.read_csv("salidasimu.hst", skiprows=6, skipfooter=7, delimiter="\s+", names=namesHST)
```

```python
hst
```

En esta siguiente figura veremos el número de partículas detectadas en función de su energía.

```python
plt.figure(figsize=(10,6))
plt.plot(hst["p_in_bin(GeV)"], hst["N_phot"], label="Fotones")
plt.plot(hst["p_in_bin(GeV)"], hst["N_e+"] + hst["N_e-"], label="Electrones y positrones")
#plt.plot(hst["p_in_bin(GeV)"], hst["N_e-"], label="N electrons")
plt.plot(hst["p_in_bin(GeV)"], hst["N_mu+"] + hst["N_mu-"], label="Muones + y -")
#plt.plot(hst["p_in_bin(GeV)"], hst["N_mu-"], label="N muons-")
plt.plot(hst["p_in_bin(GeV)"], hst["N_pi+"] + hst["N_pi-"] + hst["N_pi0"], label="Piones +, - y 0")
plt.plot(hst["p_in_bin(GeV)"], hst["N_p"] + hst["N_n"], label="Protones y neutrones")
plt.plot(hst["p_in_bin(GeV)"], hst["Total_per_bin"], label="Total")

plt.xlabel("Energía (GeV)")
plt.ylabel("Número de partículas")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.grid()
plt.show()
```

El "charge ratio" para muones atmosféricos (el flujo de muones + sobre el flujo de muones -) se ha medido experimentalmente y es aproximadamente 1.26 (Agafonova et al. 2015). Para nuestra simulación, este ratio fue de:

```python
print("Razón mu+ / mu- es", hst["N_mu+"].sum() / hst["N_mu-"].sum())
```

Finalmente observaremos el archivo salidasimu.shw.bz2, que es un comprimido con la información de todas las partículas secundarias de la simulación. En este archivo, se lista cada partícula secundaria individualmente y se muestran sus propiedades, como el moméntum y su posición de detección

```python
namesSHW = ['CorsikaId','px','py','pz','x','y','t','shower_id','prm_id',
      'prm_energy','prm_theta','prm_phi']
shw = pd.read_csv("salidasimu.shw.bz2", skiprows=5, skipfooter=1, delimiter="\s+", names=namesSHW)
```

```python
shw
```

A esta tabla le añadiremos cuatro columnas más:

p, que mostrará el moméntum total, sqrt(px^2+py^2+pz^2)

r, que mostrará la distancia de detección desde el centro de la lluvia, sqrt(x^2+y^2)

theta, el ángulo de incidencia o cenital, arccos(pz/p)

phi, el ángulo azimutal, arctan(py,px)

```python
shw["p"] = np.sqrt(shw["px"] ** 2 + shw["py"] ** 2 + shw["pz"] ** 2)
shw["r"] = np.sqrt(shw["x"] ** 2 + shw["y"] ** 2)
shw["theta"] = np.arccos(shw["pz"] / shw["p"])
shw["phi"] = np.arctan(shw["py"] / shw["px"])
```

```python
shw
```

Las siguientes dos figuras son histogramas donde, a partir de los datos de shw, mostramos el número de partículas de cada tipo respecto de su distancia de detección y su moméntum. Esto significa entonces que estos histogramas deben ser similares a las dos primeras figuras que hicimos, lo cual corroboraría que la información en los tres archivos (el .dst, el .hst y el .shw.bz2) es la misma.

```python
plt.figure(figsize=(10,6))

shw[shw["CorsikaId"]==1]["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="Fotones")
shw[(shw["CorsikaId"]==2)|(shw["CorsikaId"]==3)]["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="e+ y e-")
shw[(shw["CorsikaId"]==5)|(shw["CorsikaId"]==6)]["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="mu+ y mu-")
shw[(shw["CorsikaId"]==7)|(shw["CorsikaId"]==8)|(shw["CorsikaId"]==9)]["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="Piones +, - y 0")
shw[(shw["CorsikaId"]==13)|(shw["CorsikaId"]==14)]["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="Protones y neutrones")
shw["r"].hist(bins=np.logspace(np.log(shw["r"].min()), np.log(shw["r"].max()), 200), histtype='step', label="Total")

plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Distancia radial (m)")
plt.ylabel("Número de partículas")
plt.xlim(1e-1, 1e5)
plt.show()
```

```python
plt.figure(figsize=(10,6))


shw[shw["CorsikaId"]==1]["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="Fotones")
shw[(shw["CorsikaId"]==2)|(shw["CorsikaId"]==3)]["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="e+ y e-")
shw[(shw["CorsikaId"]==5)|(shw["CorsikaId"]==6)]["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="mu+ y mu-")
shw[(shw["CorsikaId"]==7)|(shw["CorsikaId"]==8)|(shw["CorsikaId"]==9)]["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="Piones +, - y 0")
shw[(shw["CorsikaId"]==13)|(shw["CorsikaId"]==14)]["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="Protones y neutrones")
shw["p"].hist(bins=np.logspace(np.log(shw["p"].min()), np.log(shw["p"].max()), 200), histtype='step', label="Total")


plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Moméntum")
plt.ylabel("Número de partículas")
plt.xlim(1e-5, 1e3)
plt.show()
```
